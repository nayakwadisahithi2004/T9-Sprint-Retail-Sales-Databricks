```sql
-- Row count reconciliation
SELECT 'silver_customers' table_name, COUNT(*) cnt FROM retail_etl.silver_customers
UNION ALL
SELECT 'silver_products', COUNT(*) FROM retail_etl.silver_products
UNION ALL
SELECT 'silver_stores', COUNT(*) FROM retail_etl.silver_stores
UNION ALL
SELECT 'silver_sales', COUNT(*) FROM retail_etl.silver_sales
UNION ALL
SELECT 'fact_sales', COUNT(*) FROM retail_etl.fact_sales
UNION ALL
SELECT 'rejected_products', COUNT(*) FROM retail_etl.rejected_products
UNION ALL
SELECT 'rejected_sales_qty', COUNT(*) FROM retail_etl.rejected_sales_qty;

-- Regional Sales Report
SELECT
    ds.Region,
    COUNT(*) AS TotalTransactions,
    SUM(fs.Quantity) AS TotalQuantity,
    SUM(fs.SalesAmount) AS TotalRevenue,
    ROUND(AVG(fs.SalesAmount),2) AS AvgOrderValue
FROM retail_etl.fact_sales fs
JOIN retail_etl.dim_store ds
ON fs.StoreID = ds.StoreID
GROUP BY ds.Region
ORDER BY TotalRevenue DESC;

-- Top Products
SELECT
    dp.ProductName,
    dp.Category,
    SUM(fs.Quantity) AS UnitsSold,
    SUM(fs.SalesAmount) AS Revenue
FROM retail_etl.fact_sales fs
JOIN retail_etl.dim_product dp
ON fs.ProductID = dp.ProductID
GROUP BY dp.ProductName, dp.Category
ORDER BY Revenue DESC
LIMIT 10;

-- Top Customers
SELECT
    dc.CustomerName,
    dc.City,
    SUM(fs.SalesAmount) AS LifetimeValue
FROM retail_etl.fact_sales fs
JOIN retail_etl.dim_customer dc
ON fs.CustomerID = dc.CustomerID
GROUP BY dc.CustomerName, dc.City
ORDER BY LifetimeValue DESC
LIMIT 10;
```
