```sql
CREATE SCHEMA IF NOT EXISTS training_catalog.bronze;
USE training_catalog.bronze;

CREATE TABLE IF NOT EXISTS customers_raw
USING CSV
OPTIONS (
  header = 'true',
  inferSchema = 'true'
)
LOCATION 's3://sales-analytics-lakehouse/sftp/customers/';
```

```sql
CREATE TABLE IF NOT EXISTS products_raw
USING CSV
OPTIONS (
  header = 'true',
  inferSchema = 'true'
)
LOCATION 's3://sales-analytics-lakehouse/sftp/products/';
```

```sql
CREATE TABLE IF NOT EXISTS stores_raw
USING CSV
OPTIONS (
  header = 'true',
  inferSchema = 'true'
)
LOCATION 's3://sales-analytics-lakehouse/sftp/stores/';
```

```sql
CREATE TABLE IF NOT EXISTS sales_raw
USING CSV
OPTIONS (
  header = 'true',
  inferSchema = 'true'
)
LOCATION 's3://sales-analytics-lakehouse/sftp/sales/';
```

```sql
SELECT 'customers' AS table_name, COUNT(*) AS row_count FROM customers_raw
UNION ALL
SELECT 'products',  COUNT(*) FROM products_raw
UNION ALL
SELECT 'stores',    COUNT(*) FROM stores_raw
UNION ALL
SELECT 'sales',     COUNT(*) FROM sales_raw;
```
