```sql
CREATE SCHEMA IF NOT EXISTS retail_etl;

SHOW SCHEMAS;

USE retail_etl;
```

```sql
DROP TABLE IF EXISTS retail_etl.bronze_customers;

CREATE TABLE retail_etl.bronze_customers
USING CSV
OPTIONS (
  path 's3://sales-analytics-lakehouse/sftp/customers/',
  header 'true',
  inferSchema 'true'
);
```

```sql
DROP TABLE IF EXISTS retail_etl.bronze_products;

CREATE TABLE retail_etl.bronze_products
USING CSV
OPTIONS (
  path 's3://sales-analytics-lakehouse/sftp/products/',
  header 'true',
  inferSchema 'true'
);
```

```sql
DROP TABLE IF EXISTS retail_etl.bronze_stores;

CREATE TABLE retail_etl.bronze_stores
USING CSV
OPTIONS (
  path 's3://sales-analytics-lakehouse/sftp/stores/',
  header 'true',
  inferSchema 'true'
);
```

```sql

DROP TABLE IF EXISTS retail_etl.bronze_sales;

CREATE TABLE retail_etl.bronze_sales
USING CSV
OPTIONS (
  path 's3://sales-analytics-lakehouse/sftp/sales/',
  header 'true',
  inferSchema 'true'
);
```

```sql
SELECT 'customers' AS table_name, COUNT(*) AS cnt FROM retail_etl.bronze_customers
UNION ALL
SELECT 'products', COUNT(*) FROM retail_etl.bronze_products
UNION ALL
SELECT 'stores', COUNT(*) FROM retail_etl.bronze_stores
UNION ALL
SELECT 'sales', COUNT(*) FROM retail_etl.bronze_sales;
```
