```sql
SELECT 'silver_customers' AS table_name, COUNT(*) AS cnt FROM retail_etl.silver_customers
UNION ALL
SELECT 'silver_products', COUNT(*) FROM retail_etl.silver_products
UNION ALL
SELECT 'silver_stores', COUNT(*) FROM retail_etl.silver_stores
UNION ALL
SELECT 'silver_sales', COUNT(*) FROM retail_etl.silver_sales
UNION ALL
SELECT 'rejected_products', COUNT(*) FROM retail_etl.rejected_products
UNION ALL
SELECT 'rejected_sales_qty', COUNT(*) FROM retail_etl.rejected_sales_qty
UNION ALL
SELECT 'fact_sales', COUNT(*) FROM retail_etl.fact_sales;
```
