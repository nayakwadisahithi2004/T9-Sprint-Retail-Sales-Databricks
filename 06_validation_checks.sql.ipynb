-- VALIDATION LAYER


-- Verify ETL correctness after Bronze, Silver, Reject and Gold processing

-- Checks reconciliation, duplicates, nulls, referential integrity,

-- final row counts and audit completeness

```sql
-- Validate Bronze vs Silver vs Reject row reconciliation
SELECT
'Customers' AS TableName,
(SELECT COUNT(*) FROM training_catalog.bronze.customers_raw) AS BronzeCount,
(SELECT COUNT(*) FROM training_catalog.silver.silver_customers) AS SilverCount,
(SELECT COUNT(*) FROM training_catalog.reject.rejected_customers) AS RejectCount

UNION ALL

SELECT
'Products',
(SELECT COUNT(*) FROM training_catalog.bronze.products_raw),
(SELECT COUNT(*) FROM training_catalog.silver.silver_products),
(SELECT COUNT(*) FROM training_catalog.reject.rejected_products)

UNION ALL

SELECT
'Stores',
(SELECT COUNT(*) FROM training_catalog.bronze.stores_raw),
(SELECT COUNT(*) FROM training_catalog.silver.silver_stores),
(SELECT COUNT(*) FROM training_catalog.reject.rejected_stores)

UNION ALL

SELECT
'Sales',
(SELECT COUNT(*) FROM training_catalog.bronze.sales_raw),
(SELECT COUNT(*) FROM training_catalog.silver.silver_sales),
(SELECT COUNT(*) FROM training_catalog.reject.rejected_sales);
```

```sql
-- Check duplicate transactions in FactSales
SELECT
TransactionID,
COUNT(*) AS cnt
FROM training_catalog.gold.FactSales
GROUP BY TransactionID
HAVING COUNT(*) > 1;
```

```sql
SELECT *
FROM training_catalog.gold.DimCustomer
WHERE CustomerID IS NULL
   OR CustomerName IS NULL
   OR Email IS NULL;

SELECT *
FROM training_catalog.gold.DimProduct
WHERE ProductID IS NULL
   OR ProductName IS NULL;

SELECT *
FROM training_catalog.gold.DimStore
WHERE StoreID IS NULL
   OR StoreName IS NULL;
```

```sql
-- Validate surrogate key lookups in FactSales
SELECT *
FROM training_catalog.gold.FactSales
WHERE CustomerSK IS NULL
   OR ProductSK IS NULL
   OR StoreSK IS NULL;
```

```sql
-- Validate final Gold table row counts
SELECT 'DimCustomer' AS table_name, COUNT(*) FROM training_catalog.gold.DimCustomer
UNION ALL
SELECT 'DimProduct', COUNT(*) FROM training_catalog.gold.DimProduct
UNION ALL
SELECT 'DimStore', COUNT(*) FROM training_catalog.gold.DimStore
UNION ALL
SELECT 'FactSales', COUNT(*) FROM training_catalog.gold.FactSales;
```

```sql
-- Validate audit records created for each ETL layer
SELECT
Layer,
COUNT(*) AS LogCount
FROM training_catalog.audit.etl_run_log
GROUP BY Layer;
```
